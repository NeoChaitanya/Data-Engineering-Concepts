```python
import pandas as pd
```

```python
df1 = pd.read_csv('student.csv')
df2 = pd.read_csv('mark.csv')
```

```python
merged_df = pd.merge(df1,df2, on='Student_id')
```

```python
left_join_df = pd.merge(df1,df2, on='Student_id', how='left')
```

```python
left_join_df
```

```python
right_join_df = pd.merge(df1,df2, on='Student_id', how='right')
```

```python
right_join_df
```

```python
outer_join_df = pd.merge(df1,df2, on='Student_id', how='outer')
```

```python
outer_join_df
```

```python
inner_join_df = pd.merge(df1,df2, on='Student_id', how='inner')
```
